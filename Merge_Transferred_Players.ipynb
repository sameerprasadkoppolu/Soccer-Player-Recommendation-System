##Now, There Could Be Players Who Were Transferred Between Europe’s Top 5 Leagues and Liga Primeira and Eredivisie. These Players Need to Be Merged.

```python
import pandas as pd
import io
from google.colab import files
uploaded=files.upload()
df1=pd.read_csv(io.BytesIO(uploaded['Outfielders_Final_EuropeTop5.csv']))
```

```python
uploaded=files.upload()
df2=pd.read_csv(io.BytesIO(uploaded['Outfielders_Liga_Portugal.csv']))
```

```python
uploaded=files.upload()
df3=pd.read_csv(io.BytesIO(uploaded['Outfielders_Eredivisie.csv']))
```

##Concatenate all 3 Dataframes

```python
df = pd.concat([df1, df2, df3], ignore_index = True)
```

```python
pd.set_option('display.max_columns', 500)
```

```python
df
```

##Merging Transferred Players Follows the Same Process. Hence, We Again Create a Custom Primary Key

```python
df['temp'] = df['Player'].map(str) + df['Nation'].map(str) + df['Age'].map(str) + df['Born'].map(str)
df.head()
```

##Identify the columns whose values are to be aggregated

```python
cols = list(df.columns)[4:-3]
```

##Identify the Unique Players using the Custom Primary Key

```python
set_temp = list(set(df['temp']))
```

```python
set_temp
```

```python
print(list(df.columns))
```

## Now we Perform the Merging

```python
X = pd.DataFrame(columns = list(df.columns))

for i in set_temp:
  t = df.loc[df['temp'] == i]

  if len(t) == 1:
    X.loc[len(X)] = list(t.values[0])
  
  if len(t) > 1:
    t1 = t.groupby(['temp', 'Player', 'Nation', 'Age', 'Born'])[cols].sum().reset_index()
    t1_values = list(t1.values[0])[5:]
    t1_squad = ' + '.join(list(t['Squad']))
    t1_temp = i
    t1_player = t1.loc[0, 'Player']
    t1_nation = t1.loc[0, 'Nation']
    t1_age = t1.loc[0, 'Age']
    t1_born = t1.loc[0, 'Born']

    #to find position:
    t2 = t.groupby(['Pos'])['Min_Playing'].sum().reset_index()
    pos_list = list(t2['Pos'])
    mins = list(t2['Min_Playing'])
    max_mins_index = mins.index(max(mins))
    pos = pos_list[max_mins_index]

    #Append to X:
    X.loc[len(X)] = [t1_player, t1_nation, t1_age, t1_born] + t1_values + [pos, t1_squad, t1_temp]
```

```python
X.drop(columns = ['temp'], inplace = True)
```
