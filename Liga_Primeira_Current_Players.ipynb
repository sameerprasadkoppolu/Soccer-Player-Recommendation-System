##Like We Did in Europe’s Top 5 Leagues, We Identify the Position of the Player Based on the Position That They Played the Maximum Amount of Time In

```python
#Find Position

for i in range(len(y)):
  t = y.loc[i, 'temp']

  df_t = df1.loc[df1['temp'] == t, ['Pos', 'Min_Playing_Time']].groupby(['Pos'])['Min_Playing_Time'].sum().reset_index()
  if len(df_t) > 1:
    print(df_t, i)
  pos = list(df_t['Pos'])
  mins = list(df_t['Min_Playing_Time'])
  max_mins_index = mins.index(max(mins))
  y.loc[i, 'Pos'] = pos[max_mins_index]
```

## We Also Find the Squad of Each Player and Mention Both Clubs a Player Has Played for if He Was Transferred During the Season

```python
#Find Squad
for i in range(len(y)):
  t = y.loc[i, 'temp']

  df_t = df1.loc[df1['temp'] == t, ['Squad']]
  squad_list = list(df_t['Squad'])

  if len(squad_list) > 1:
    print(squad_list, i)

  s = ' + '.join(squad_list)

  y.loc[i, 'Squad'] = s
```
