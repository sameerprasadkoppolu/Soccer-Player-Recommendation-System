##Identify the year of birth for each player

```python
#Calculate Year of Birth for DF1
from math import floor, ceil


df1['Born'] = 0

for i in range(len(df1)):
  print(df1.loc[i, 'Age'])
  age_str = str(df1.loc[i, 'Age']).split('-')
  age = float(age_str[0]) + (float(age_str[1]) / 365.0)
  age = round(age, 1)
  if ceil(age) - age <= 0.5:
    age = ceil(age)
  else:
    age = floor(age)
  born = 2023 - age
  df1.loc[i, 'Born'] = born
```

##Now we Merge the Transferred Players

```python
y = y.groupby(['temp', 'Player', 'Nation', 'Age', 'Born'])[cols].sum().reset_index()
```

##Identify the Position

```python
#Find Position

for i in range(len(y)):
  t = y.loc[i, 'temp']

  df_t = df1.loc[df1['temp'] == t, ['Pos', 'Min_Playing_Time']].groupby(['Pos'])['Min_Playing_Time'].sum().reset_index()
  if len(df_t) > 1:
    print(df_t, i)
  pos = list(df_t['Pos'])
  mins = list(df_t['Min_Playing_Time'])
  max_mins_index = mins.index(max(mins))
  y.loc[i, 'Pos'] = pos[max_mins_index]
```

##Identify the Squad. A player who as been transferred will have both squads listed under the 'Squad' variable

```python
#Find Squad
for i in range(len(y)):
  t = y.loc[i, 'temp']

  df_t = df1.loc[df1['temp'] == t, ['Squad']]
  squad_list = list(df_t['Squad'])

  if len(squad_list) > 1:
    print(squad_list, i)

  s = ' + '.join(squad_list)

  y.loc[i, 'Squad'] = s
```
