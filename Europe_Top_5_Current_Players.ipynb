##Now, There Are Some Players Who Are Transferred in the January Transfer Window. Since We Want Each Row to Represent a Unique Player, We Need to Sum Up the Stats of the Transferred Players. This Is Where Merging Comes Into Play and We Do So by Creating a Primary Key Called ‘Temp’ Which Contatenates the Values of ‘Player’, ‘Nation’, ‘Age’, and ‘Born’ (i.e. Birth Year)

```python
df['temp'] = df['Player'].map(str) + df['Nation'].map(str) + df['Age'].map(str) + df['Born'].map(str)
```

```python
df.head()
```

```python
#Create a Copy of the Dataframe
y = df.copy()
```

```python
print(list(df.columns))
```

```python
#Identify the Columns whose values are to be summed up for transferred players
cols = list(df.columns)[8:-1]
cols
```

##Now We Perfrom the Merging Using the Groupby() Function

```python
y = y.groupby(['temp', 'Player', 'Nation', 'Age', 'Born'])[cols].sum().reset_index()
y
```

## Now After Performing the Merge, We Need to Identify the Position of the Players. For a Transferred Player, the Position of the Player Is Chosen by Identifying the Position That the Player Played the Maximum Time In

```python
#Find Position

for i in range(len(y)):
  t = y.loc[i, 'temp']

  df_t = df.loc[df['temp'] == t, ['Pos', 'Min_Playing']].groupby(['Pos'])['Min_Playing'].sum().reset_index()
  pos = list(df_t['Pos'])
  mins = list(df_t['Min_Playing'])
  max_mins_index = mins.index(max(mins))
  y.loc[i, 'Pos'] = pos[max_mins_index]
```

## We Also Identify the Squads That the Player Has Played For. If a Player Was Transferred, Then the ‘Squad’ Variable Will Contain Both Teams That the Player Has Played for Separated by a ‘+’ Symbol.

```python
#Find Squad

for i in range(len(y)):
  t = y.loc[i, 'temp']

  l = df.loc[df['temp'] == t, ['Squad']]
  l_list = list(l['Squad'])

  s = ''

  if len(l_list) == 1:
    s = l_list[0]
  else:
    s = l_list[0] + ' + ' + l_list[1]

  y.loc[i, 'Squad'] = s
```
